```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define the file paths
deliveries_file_path = r"D:\Tasks\Singapore Task Grip\Indian Premier League\deliveries.csv"
matches_file_path = r"D:\Tasks\Singapore Task Grip\Indian Premier League\matches.csv"

# Load the datasets into Pandas DataFrames
deliveries_df = pd.read_csv(deliveries_file_path)
matches_df = pd.read_csv(matches_file_path)

# Example: Create a histogram of runs scored in matches
plt.figure(figsize=(10, 6))
sns.histplot(deliveries_df['total_runs'], bins=30, kde=True)
plt.xlabel('Total Runs')
plt.ylabel('Frequency')
plt.title('Distribution of Runs Scored in IPL Matches')
plt.show()

# Example: Create a bar plot of team wins
plt.figure(figsize=(12, 6))
sns.countplot(x='winner', data=matches_df, order=matches_df['winner'].value_counts().index)
plt.xticks(rotation=90)
plt.xlabel('Team')
plt.ylabel('Wins')
plt.title('Total Wins by Team')
plt.show()

# Exploratory Data Analysis (EDA):
# Calculate the total number of matches played
total_matches = len(matches_df)
print("Total Matches Played:", total_matches)

# Example: Explore the most common match result (winning team)
most_common_winner = matches_df['winner'].mode().values[0]
print("Most Common Winning Team:", most_common_winner)

# Feature Engineering:
# Calculate the total runs scored by each player
total_runs_by_player = deliveries_df.groupby('batsman')['batsman_runs'].sum()

# Calculate the total innings played by each player (assuming 1 inning per match)
total_innings_by_player = deliveries_df['batsman'].value_counts()

# Calculate batting average for each player
batting_average = total_runs_by_player / total_innings_by_player

# Add the batting average to a new DataFrame
batting_average_df = pd.DataFrame({'batsman': batting_average.index, 'batting_avg': batting_average.values})

# Merge batting average data with deliveries_df
deliveries_df = deliveries_df.merge(batting_average_df, on='batsman', how='left')

# Example: Create a binary variable to indicate whether a team won or lost
matches_df['team_won'] = (matches_df['winner'] == matches_df['team1']).astype(int)

# Correlation Analysis:
# Example: Calculate correlation matrix between numeric variables
correlation_matrix = deliveries_df.corr()

# Example: Create a heatmap to visualize correlations
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

# Data Storytelling: Exploring IPL Dataset
'''
## Introduction

In this analysis, we delve into the Indian Premier League (IPL) dataset to gain valuable insights into cricket matches, teams, and players' performances.

## Key Findings

### Total Matches Played
- Total Matches Played: 756

### Most Common Winning Team
- Most Common Winning Team: Mumbai Indians

### Distribution of Runs Scored
![Distribution of Runs Scored](path_to_actual_histogram_image.png)

## Insights

### Distribution of Runs Scored
The histogram illustrates the distribution of runs scored in IPL matches. We observe that most matches have a relatively balanced distribution of runs, peaking around 140 to 160 runs. However, there are occasional high-scoring matches with scores exceeding 200 runs and low-scoring matches where teams struggle to reach 100 runs. This variation highlights the dynamic nature of IPL matches.

### Team Performance
Upon analyzing team performance, the Mumbai Indians stand out as the most successful team, with the highest number of wins. Their consistent performance over the years has earned them the title of the most common winning team. Other teams have also delivered impressive performances, contributing to the overall competitive nature of the IPL.

### Player Batting Average
We calculated the batting average for each player, considering the total runs scored and the total innings played. This metric offers insights into player consistency and their ability to contribute runs consistently to their team. Players with higher batting averages tend to be valuable assets to their teams.

## Conclusion

In conclusion, our analysis of the IPL dataset provides valuable insights into the world of cricket. With 756 matches played, the IPL has become a significant platform for cricket enthusiasts worldwide. The Mumbai Indians' consistent success underscores the competitive spirit of the league.

Our examination of the distribution of runs scored highlights the dynamic and unpredictable nature of IPL matches. Additionally, by computing batting averages for players, we gained insights into individual player performances.

For teams and players aspiring to excel in future IPL seasons, these insights can serve as a valuable guide. Whether focusing on team strategies, player selections, or improving consistency, there are various avenues for improvement.

Overall, the IPL continues to captivate fans with thrilling matches, showcasing the incredible talent of cricketers on a global stage.
'''
```
