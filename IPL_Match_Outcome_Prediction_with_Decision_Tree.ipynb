```python
# Import necessary libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
import pydotplus  # Add this import statement
from IPython.display import Image
import os

# Replace 'path_to_dot_exe' with the actual path to dot.exe
os.environ["PATH"] += os.pathsep + 'C:\\Users\\pc\\Downloads\\Applications\\windows_10_msbuild_Release_graphviz-9.0.0-win32\\Graphviz\\bin'

# Load the dataset from your local file
file_path = "D:/Tasks/Singapore Task Grip/Task 05/Iris.csv"  # Update with your actual file path
data = pd.read_csv(file_path)

# Replace 'Species' with the actual name of your target variable column
X = data.drop('Species', axis=1)  # Features
y = data['Species']  # Target variable

# Create a Decision Tree classifier
clf = DecisionTreeClassifier()
clf.fit(X, y)

# Visualize the Decision Tree
dot_data = export_graphviz(clf, out_file=None, 
                            feature_names=X.columns,  
                            class_names=y.unique(),
                            filled=True, rounded=True,  
                            special_characters=True)  
graph = pydotplus.graph_from_dot_data(dot_data)
# Save the graph as 'graph.png'
graph.write_png('graph.png')

# Display the saved graph
Image('graph.png')
```
